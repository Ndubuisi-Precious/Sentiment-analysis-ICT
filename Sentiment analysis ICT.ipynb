We scrapped data from Nairaland a Nigeria forum and community whee wide range of topics and disccussions including news, business, entertainment sports, politics and more.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_nairaland_comments(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        comment_elements = soup.find_all('td', class_='l w pd')

        comments = []

        for comment in comment_elements:
            comment_text = comment.find('div', class_='narrow')
            if comment_text:
                comments.append(comment_text.text)
        
        # Create a DataFrame from the list of comments
        df = pd.DataFrame({'Comments': comments})

        # Save the DataFrame to a CSV file
        df.to_csv('nairaland_comments.csv', index=False)

        print("Comments saved to 'nairaland_comments.csv'")
    else:
        print(f"Failed to fetch page. Status code: {response.status_code}")

# Replace 'your_nairaland_url' with the actual URL you want to scrape comments from
scrape_nairaland_comments('https://www.nairaland.com/6325629/covid-19-second-wave-all-schools')
```

## Data Preprocessing
I can't achieve scucess without going the extra mile of cleaning the sucess column. I decided to remove unwanted character/symbols, drop some duplicate, corrected name misspell

```python
import neattext.functions as nfx
import re
import string
from nltk import word_tokenize
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
ict_df["Comments"] = ict_df['Comments'].astype(str)
```

```python
def data_processing(text):
    text = text.lower()
    text = re.sub(r'https\S+|WWW\S+https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\@W+|\#','',text)
    text = re.sub(r'[^\W\S]','',text)
    text = re.sub(r'""','',text)
    text = re.sub(r"'",'',text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]
    return " ".join(filtered_text)
```

```python
ict_df['cleaned_text'] = ict_df['Comments'].apply(data_processing)

ict_df.head(4)
```

```python
#### cleaning text remove multiple spaces, emojies, and special characters
ict_df['cleaned_text'] = ict_df['cleaned_text'].apply(nfx.remove_stopwords)
ict_df['cleaned_text'] = ict_df['cleaned_text'].apply(nfx.remove_multiple_spaces)
ict_df['cleaned_text'] = ict_df['cleaned_text'].apply(nfx.remove_puncts)
ict_df['cleaned_text'] = ict_df['cleaned_text'].apply(nfx.remove_punctuations)
ict_df['cleaned_text'] = ict_df['cleaned_text'].apply(nfx.remove_special_characters)
ict_df['cleaned_text'] = ict_df['cleaned_text'].apply(nfx.remove_emojis)
ict_df['cleaned_text'] = ict_df['cleaned_text'].apply(nfx.remove_punctuations)
ict_df['cleaned_text'] = ict_df['cleaned_text'].apply(nfx.remove_html_tags)

ict_df.head(20)
```

```python
# To avoid name repetitions I replace names
def correctBankNames(cleaned_text):
    replacements = [('covid19', 'covid'), ('students', 'student'), ('schools', 'school'), ('covid1919', 'covid'),
                   ('kids', 'children'), ('nigerian', 'nigeria')]
    for pat, repl in replacements:
        cleaned_text= re.sub(pat, repl, cleaned_text)
    return cleaned_text
```

## Sentiment Analysis
I created a polarity to ascertain the score of every tweet from negative, positive to neutral, plotted a bar and pie chart customers perception

```python
# I decided to use Stemmer to remove suffixs
stemmer = PorterStemmer()
def stemming(data):
    text = [stemmer.stem(word) for word in data]
    return data
```

```python
ict_df['cleaned_text'] = ict_df['cleaned_text'].apply(lambda x: stemming(x)) #apply our function
```

```python
#To obtain polarity score, i created this function TextBlob
from textblob import TextBlob

def polarity(cleaned_text):
    return TextBlob(cleaned_text).sentiment.polarity
```

```python
ict_df['polarity'] = ict_df['cleaned_text'].apply(polarity) #I applied the function and created a new column
ict_df.head(10)
```

```python
#To obtain sentiment score, I created this function TextBlob
def sentiment(label):
    if label <0:
        return "NEGATIVE"
    elif label ==0:
        return "Neutral"
    elif label >0:
        return 'POSITIVE'
```

```python
ict_df['sentiment'] = ict_df["polarity"].apply(sentiment) #We applied the function and create new column
ict_df.tail()
```

## Classification Model

```python
#imported sklearn library for classifications and predictive data analysis
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

## Bag of words vectorization

```python
vectorizer = CountVectorizer(ngram_range=(1,2)).fit(ict_df['cleaned_text'])
feature_names = vectorizer.get_feature_names()
```

```python
x = ict_df['cleaned_text']
y = ict_df['label']
x = vectorizer.transform(x)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33, random_state=42)
```

```python
print('Size of x_train', (x_train.shape))
print('Size of y_train', (y_train.shape))
print('Size of x_test', (x_test.shape))
print('Size of y_test', (y_test.shape))
```

### Classification
####  Logistics Regression

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
logreg_pred = logreg.predict(x_test)
logreg_acc = accuracy_score(logreg_pred, y_test)
print('Test accuracy: {:.2f}%'.format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test, logreg_pred))
print('\n')
print(classification_report(y_test, logreg_pred))
```

```python
# Confuison matrix to allow visualizatin of the performance of af an algorithm.
from matplotlib import style
style.use('classic')
cm = confusion_matrix(y_test, logreg_pred, labels=logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=logreg.classes_)
disp.plot()
```

### Tuning our model (with Grid Search)

```python
from sklearn.model_selection import GridSearchCV


param_grid = {'C':[0.001, 0.01, 0.1, 10]}
grid = GridSearchCV(LogisticRegression(), param_grid)
grid.fit(x_train, y_train)
```

```python
print('Best parameter:', grid.best_params_)
```

```python
y_pred = grid.predict(x_test)
grid_accura = accuracy_score(y_pred, y_test)
print('Test accuracy: {:.2f}%'.format(grid_accura*100))
```

```python
print(confusion_matrix(y_test, y_pred))
print('\n')
print(classification_report(y_test, y_pred))
```

### Linear SVM (support vector machine)

```python
from sklearn.svm import LinearSVC

svc_model = LinearSVC()
svc_model.fit(x_train, y_train)
```

```python
svc_pred = svc_model.predict(x_test)
svc_acc = accuracy_score(svc_pred, y_test)
print('Test accuracy: {:.2f}%'.format(svc_acc*100))
```

```python
print(confusion_matrix(y_test,svc_pred))
print('\n')
print(classification_report(y_test, svc_pred))
```
